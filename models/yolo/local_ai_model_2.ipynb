```python

from ultralytics import YOLO

# Load a pretrained YOLOv10n model
model = YOLO("../weight/yolo.pt")
```

```python
# 카메라 켜기
cap = cv2.VideoCapture(0)
# 웹캠 오류 처리
if not cap.isOpened():
    print("camera open failed")
    exit()
```

```python
# 비디오 저장 설정
output_filename = '../result/video/output_result2.mp4'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
fps = 20.0
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_filename, fourcc, fps, (frame_width, frame_height))
```

```python
exit_flag = False
try:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # YOLO 추론
        results = model(frame)

        # 결과 프레임 (Bounding Box 포함)
        annotated_frame = results[0].plot()

        # 화면 출력
        cv2.imshow('YOLO Detection', annotated_frame)

        # 파일 저장
        out.write(annotated_frame)
        
        # 박스 좌표 출력
        for result in results:
            for box in result.boxes: #.boxs는 ultralytics에서 제공하는 커스텀 클래스
                x1, y1, x2, y2 = box.xyxy[0].tolist()
                conf = box.conf[0].item()
                cls = box.cls[0].item()
                # print(f"Class: {cls}, Confidence: {conf:.2f}, Box: [{x1:.1f}, {y1:.1f}, {x2:.1f}, {y2:.1f}]")
                # 종료 키
                if cv2.waitKey(10) & 0xFF == ord('q'):
                    exit_flag = True
                    break
            if exit_flag == True:
                break

        # 결과 화면 출력
        annotated = results[0].plot()
        cv2.imshow('YOLO Detection', annotated)

        if exit_flag == True:
                break
        # # 종료 키
        # if cv2.waitKey(10) & 0xFF == ord('q'):
        #     break
finally:        
    # 자원 해제
    cap.release()
    out.release()
    cv2.destroyWindow("YOLO Detection")
```
